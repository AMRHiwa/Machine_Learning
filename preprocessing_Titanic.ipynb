```python
def f(name):
    if '.' in name:
        return name.split(',')[1].split('.')[0].strip()
    else:
        return 'Unknow'
```

```python
def g(t):
    if t == 'Mr':
        return 1
    elif t == 'Master':
        return 3
    elif t in ['Ms', 'Mlle', 'Miss']:
        return 4
    elif t in ['Mrs', 'Mme']:
        return 5
    else:
        return 2
```

```python
df['title'] = df['Name'].apply(f).apply(g)
df.head(10)
```

```python
t = pd.crosstab(df['title'], df['Survived'])
t
```

```python
t_pict = t.div(t.sum(1).astype(float), axis=0)
t_pict
```

```python
t_pict.plot(kind='bar', stacked=True, title='Survival Rate by title')
plt.xlabel('title')
plt.ylabel('Survival Rate')
```

# Example GroupBy

```python
df = pd.DataFrame({'Bird': list('AABBB'), 'Speed': [380, 370, 24, 26, np.nan]})
df
```

```python
df.groupby('Bird').mean()
```

```python
df['Speed'] = df.groupby(['Bird'])['Speed'].transform(lambda x : x.fillna(x.mean()))

df
```
