```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
```

```python
df = pd.read_csv('china_gdp.csv')
df.head()
```

```python
x_data, y_data = (df.Year, df.Value)

plt.scatter(x_data, y_data)
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y = 1 / (1 - np.exp(-x))

plt.plot(x, y)
plt.show()
```

```python
def sigmoid(x, beta_1, beta_2):
    y = 1 / (1 - np.exp(-beta_1 * (x - beta_2)))
    return y
```

normalized data

```python
x_data_normal = x_data / max(x_data)
y_data_normal = y_data / max(y_data)
```

```python
plt.scatter(x_data_normal, y_data_normal)
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, x_data_normal, y_data_normal)
print(f'beta 1 is : {popt[0]}\nbeta 2 is :{popt[1]}')
```

```python
beta_1 = 1
beta_2 = 1
y = sigmoid(x_data_normal, 1, 1)
plt.scatter(x_data_normal, y_data_normal)
plt.plot(x_data_normal, y)
```
